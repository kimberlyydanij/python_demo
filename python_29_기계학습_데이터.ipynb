- StandardScaler 
    - 평균 0, 표준편차 1
    - $z = \frac{x-\mu}{\sigma}$, $\mu$는 평균 $\sigma$는 표준편차
- MinMaxScaler
    - 최솟값 0, 최댓값 1
    - $z = \frac{x-x_{min}}{x_{max}- x_{min}}$

### MinMaxScaler


```python
# stratify는 데이터의 비율을 동일하게 나누는 옵션이다.
X_train, X_test, y_train, y_test = train_test_split(breast_cancer.iloc[:,:-1], breast_cancer.target, stratify=breast_cancer.target)
```

```python
mm = MinMaxScaler()
mm.fit(X_train[['worst area']])
X_train[['worst area']] = mm.transform(X_train[['worst area']])
X_test[['worst area']] = mm.transform(X_test[['worst area']])
```

```python
lr2 = LogisticRegression()
lr2.fit(X_train, y_train)
lr2.score(X_test, y_test)
```

### one hot encoding
문자열 데이터를 column으로 바꾸고, 해당하는 위치에만 1로 한다.

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
```

```python
# return sparse matrix 이다.
# 0이 많은 희소행렬은 메모리 효율성을 위해서 압축된 형식의 sparse matrix로 만든다다
ohe.fit_transform(iris[['species']])
```

```python
#toarray()는 numpy로 바꿔줘서 값을 확인 할 수 있다.
ohe.fit_transform(iris[['species']]).toarray()
```

```python
# 다시 문자로 바꿀 수 있다.
ohe.inverse_transform([[1,0,0],[0,1,0],[0,0,1]])
```
