### Where

```python
a = np.array([1,2,3,4,5])
a
```

```python
np.where(a>3)   #인덱스값 3,4
```

```python
a = np.array([10,20,30,40,50])
a
```

```python
#where의 인자는 조건이다.
#조건만 넣을 경우 True의 위치를 리턴한다
np.where(a>30)
```

```python
# where는 인자 개수에 따라서 기능이 달라진다.
# 조건이 True 이면 0, False  이면 100을 반환한다.
np.where(a>30,0,100)
```
