### at
이름으로 한 위치의 값만 가져온다.

```python
data.at[0, 'tip']
#data.loc[0, 'tip']
#data.iloc[0, 1]
```

```python
# at은 slicing을 사용할 수 없다.
#data.at[0:3, 'tip']
```

```python
# at은 fancying indexing을 사용할 수 없다.
data.at[[1,3,5],['tip','smoker', 'day']]
```

### iat
iat은 인덱스 번호로 한 위치의 값을 가져온다.

```python
data.iat[0,1]
#data.at[0,'tip']
```

```python
#iat은 슬라이싱을 사용할 수 없다.
data.iat[0:3, 1]
```

### boolean indexing

```python
# data.tip >= 5
data['tip'] >= 5
```

```python
data[data['tip'] >= 5]  #data[True, False,True,True.....]
```

```python
data.tip
data['tip']
#a = data[data['tip']>=5]
#a['tip']

data[data['tip']>=5]['tip'] # data[data['tip']>=5]의 결과는 DataFrame
#data['tip'][3]             # data['tip']의 결과는 Series
```

```python
data[data['tip']>=5].loc[:, 'tip':'day']
```

```python
data[data['tip']>=5].iloc[:, 1:5]
```

```python
data[data['tip']>=5][['tip','day']]
```

요일이 'Sun' 이거나 'Sat'의 tip과 day의 값의 가져오기

```python
data[(data['day']=='Sun') | (data['day']=='Sat')][['tip', 'day']]

a = data[(data['day']=='Sun') | (data['day']=='Sat')]
a['day'].unique()
a[['tip', 'day']]
```
